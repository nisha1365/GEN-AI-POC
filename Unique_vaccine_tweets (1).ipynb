```python
# Create a new DataFrame with unique values from the 'text' column  
new_df = pd.DataFrame({'text': df['text'].unique()})  
```

```python
new_df
```

```python
import re
```

```python
emoji_pattern = re.compile("["  
        u"\U0001F600-\U0001F64F"  # emoticons  
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs  
        u"\U0001F680-\U0001F6FF"  # transport & map symbols  
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)  
                           "]+", flags=re.UNICODE)  
```

```python
import re  
  
def clean_text(text):  
    # Remove RT prefix from retweets  
    text = re.sub(r'^RT[\s]+', '', text)  
      
    # Remove ':' character that follows RT prefix  
    text = re.sub(r':', '', text)  
      
    # Remove mentions (@)  
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)  
      
    # Remove hashtags (#)  
    text = re.sub(r'#[A-Za-z0-9_]+', '', text)  
  
    # Remove emojis  
    text = re.sub(emoji_pattern, '', text)  
      
    return text  
  
  

```

```python
# Apply the clean_text function to the 'text' column  
new_df['clean_text'] = new_df['text'].apply(clean_text)  
```
