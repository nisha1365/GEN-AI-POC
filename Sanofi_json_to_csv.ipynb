```python

import json
import csv
import os

# Directory containing your JSON files
directory_path = "/content/sample_data/haempphilia-json-data-mars23-mai23"

# List all files in the directory
json_files = [file for file in os.listdir(directory_path) if file.endswith('.json')]

# CSV file to store the combined data
csv_file_path = "/content/combined_data.csv"

# Write the header to the CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    header_written = False

    # Iterate through each JSON file
    for file_name in json_files:
        file_path = os.path.join(directory_path, file_name)

        try:
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)

                # Check if the header has been written
                if not header_written:
                    # Write the header using the keys from the first tweet
                    header = list(data['tweet_list'][0].keys())
                    csv_writer.writerow(['file_name'] + header)
                    header_written = True

                # Write each tweet to the CSV file
                for tweet in data['tweet_list']:
                    csv_writer.writerow([file_name] + list(tweet.values()))

        except json.JSONDecodeError as e:
            print(f"Error decoding JSON in file {file_name}: {e}")

print(f"CSV file created at: {csv_file_path}")
```

## Counting Number of Tweets in each file.

```python
import csv
from collections import Counter

csv_file_path = "/content/combined_data.csv"

# Dictionary to store the count of tweets for each file
tweet_counts = Counter()

with open(csv_file_path, 'r', newline='', encoding='utf-8') as csv_file:
    csv_reader = csv.DictReader(csv_file)

    for row in csv_reader:
        file_name = row['file_name']
        tweet_counts[file_name] += 1

# Print the tweet counts for each file
for file_name, count in tweet_counts.items():
    print(f"File: {file_name}, Number of Tweets: {count}")
```

```python
import pandas as pd
from transformers import BertTokenizer
import re

# Load BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def preprocess_tweet(tweet_text):
    tweet_text = tweet_text.lower()
    tweet_text = re.sub(r'https?:\/\/[^\s]+', '', tweet_text)
    tweet_text = re.sub(r'[^a-zA-Z\s]', '', tweet_text)
    tokens = tokenizer.tokenize(tokenizer.decode(tokenizer.encode(tweet_text)))
    stopwords = ['the', 'and', 'to', 'in', 'with', 'for', 'of', 'on', 'at', 'by']
    tokens = [token for token in tokens if token not in stopwords]
    return ' '.join(tokens)



# Apply preprocessing to the 'text' column
df['cleaned_text'] = df['text'].apply(preprocess_tweet)

# Display the DataFrame with the cleaned_text column
print(df[['text', 'cleaned_text']])
```

```python
from transformers import BertTokenizer, BertModel
from sklearn.cluster import AgglomerativeClustering
import pandas as pd

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Function to get BERT embeddings
def get_bert_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation = True, padding = True, max_length=128)
    outputs = model(**inputs)
    return outputs.pooler_output.detach().numpy()



# Get BERT embeddings for each cleaned text
df['bert_embedding'] = df['cleaned_text'].apply(get_bert_embedding)

# Extract the embeddings into a list
embeddings_list = df['bert_embedding'].to_list()

# Convert the list of embeddings to a 2D array
embedding_matrix = [embedding.flatten() for embedding in embeddings_list]

# Apply hierarchical clustering
clustering = AgglomerativeClustering(n_clusters=5)  # Adjust the number of clusters as needed
labels = clustering.fit_predict(embedding_matrix)

# Add the clustering labels to the DataFrame
df['cluster_label'] = labels

# Display the DataFrame with clustering labels
print(df[['cleaned_text', 'cluster_label']])
```
