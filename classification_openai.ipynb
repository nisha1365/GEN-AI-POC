```python
new_df['label'] = 0      # 0 for non hcp
new_df['score'] = 0.0    # Initial score

# Loop through tweets and generate classifications
for index, tweet in new_df['text'].iteritems():
  prompt = (
      f"Classify whether the following tweet is from a Health Care Provider(HCP) or not.\n\n"
      f"Instructions: Healthcare Providers (HCPs) are professional working in the healthcare industry."
      f"They include doctors, nurses, medical practitioners, and other working in the healthcare industry."
      f"diagnosis, or treatment.If the tweet is from someone in a healthcare-related role, classify it as'HCP'; "
      f"otherwise, classify it as 'NON HCP'.\n\n"
      f"Tweet:\n\"{tweet}\"\n\nLabel: "

  )

  # Request classification from ChatGPT
  response = openai.ChatCompletion.create(
      model = "gpt-3.5-turbo",
      messages = [
          {"role": "system","content": "you are a helpful assistant."},
          {"role": "user", "content": prompt}
      ]

  )

  model_response = response['choices'][0]['text'].strip()
  model_score = response['choices'][0]['logprobs']['token_logprobs'][0]

  # Update Dataframe with model predictions
  if model_response =="HCP":
    new_df[index,'label'] = 1
  new_df.at[index, 'score'] = model_score


print(new_df)

```
