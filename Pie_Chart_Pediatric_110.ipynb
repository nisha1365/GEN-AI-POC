```python
import pandas as pd
topics_list = pd.read_csv("/content/topics_name_110.csv")
topics_list
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import numpy as np

# Load the SentenceTransformer model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Embed the proper topic names
proper_topic_names = topics_list['Topic_Name'].tolist()
embeddings = model.encode(proper_topic_names)

# Convert embeddings to a 2D array
embeddings = np.array(embeddings)

# Add the embeddings as a new column in the dataframe
topics_list['embeddings'] = embeddings.tolist()
```

```python
# Extract the embeddings as a numpy array
embeddings = np.array(topics_list['embeddings'].tolist())

# Get the shape of the first embedding
embedding_shape = embeddings[0].shape

# Iterate through the embeddings and check if they all have the same shape
all_same_length = all(embedding.shape == embedding_shape for embedding in embeddings)

if all_same_length:
    print("All embeddings have the same length.")
else:
    print("Embeddings have different lengths.")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Get the embeddings column as a numpy array
embeddings = np.array(topics_list['embeddings'].tolist())

# Define a range of possible cluster numbers to try
min_clusters = 2
max_clusters = 10
num_clusters_range = range(min_clusters, max_clusters+1)

# Initialize empty lists to store the results
inertias = []
silhouette_scores = []

# Loop over the range of possible cluster numbers
for k in num_clusters_range:
    # Train the K-Means model
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(embeddings)

    # Calculate the inertia and silhouette score
    inertia = kmeans.inertia_
    silhouette = silhouette_score(embeddings, kmeans.labels_)

    # Append the results to the lists
    inertias.append(inertia)
    silhouette_scores.append(silhouette)

# Plot the elbow curve to determine the optimal number of clusters
plt.plot(num_clusters_range, inertias, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()

# Plot the silhouette scores to determine the optimal number of clusters
plt.plot(num_clusters_range, silhouette_scores, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score')
plt.show()

# Determine the optimal number of clusters based on the elbow method
diffs = np.diff(inertias)
diffs_ratio = diffs[1:] / diffs[:-1]
k_elbow = num_clusters_range[np.argmax(diffs_ratio)+1]
print("Optimal number of clusters based on elbow method:", k_elbow)

# Determine the optimal number of clusters based on the silhouette score
k_silhouette = num_clusters_range[np.argmax(silhouette_scores)]
print("Optimal number of clusters based on silhouette score:", k_silhouette)

```

It is possible for the elbow method and silhouette score to give different optimal numbers of clusters. The elbow method is based on the sum of squared distances between data points and their cluster centers, while the silhouette score is based on the similarity of data points to their own cluster compared to other clusters.

In some cases, the elbow method may suggest a higher number of clusters than the silhouette score because it is based on minimizing the sum of squared distances, which tends to favor more clusters. On the other hand, the silhouette score may suggest a lower number of clusters because it takes into account the similarity of data points to their own cluster compared to other clusters, which tends to favor fewer clusters.

Ultimately, the optimal number of clusters depends on the specific dataset and the goals of the analysis. It is important to consider both the elbow method and silhouette score, as well as any other relevant factors, when determining the optimal number of clusters for a particular dataset.

```python

# Extract the embeddings as a numpy array
embeddings = np.array(topics_list['embeddings'].tolist())

# Apply K-means clustering with 8 clusters
kmeans = KMeans(n_clusters=8, random_state=42)
cluster_labels = kmeans.fit_predict(embeddings)

# Update the cluster labels in the topics_list dataframe
topics_list['cluster_labels'] = cluster_labels
```

```python
# Get the cluster counts
cluster_counts = topics_list['cluster_labels'].value_counts().sort_index()
cluster_counts
```

```python
import matplotlib.pyplot as plt

# Group the dataframe by cluster labels and count the occurrences of each proper topic name
cluster_counts = topics_list.groupby('cluster_labels')['Topic_Name'].value_counts()

# Plot a pie chart for each cluster
num_clusters = len(cluster_counts.index.levels[0])
fig, axs = plt.subplots(num_clusters, figsize=(6, 6 * num_clusters))
for i, (cluster_label, counts) in enumerate(cluster_counts.groupby(level=0)):
    axs[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    axs[i].set_title(f'Cluster {cluster_label} - Proper Topic Name Distribution')
plt.tight_layout()
plt.show()
```

## Getting All Clusters in CSV

```python
import pandas as pd

# Create a dictionary of clusters and the topics in each cluster
cluster_dict = {}
for i in range(8):
    cluster_dict[i] = list(topics_list[topics_list['cluster_labels']==i]['Topic_Name'])

# Find the length of the longest list
max_len = max(len(lst) for lst in cluster_dict.values())

# Add None values to the shorter lists to make them the same length
for i in range(8):
    while len(cluster_dict[i]) < max_len:
        cluster_dict[i].append(None)

# Create a DataFrame from the cluster dictionary
df = pd.DataFrame(cluster_dict)

# Save the DataFrame to a CSV file
df.to_csv('cluster_topics_Pediatric_110.csv', index=False)

```
