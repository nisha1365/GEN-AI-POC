```python
from googlesearch import search
from bs4 import BeautifulSoup
import requests

def get_top_links(query):
    # Search for the query and get top 5 links
    top_links = []
    for j in search(query, num=5):
        top_links.append(j)
    return top_links

def main():
    name = input("Enter the celebrity name: ")
    links = get_top_links(name)
    print(f"Top 5 links for {name}:")
    for link in links:
        print(link)

if __name__ == "__main__":
    main()
```

```python
from googlesearch import search
from bs4 import BeautifulSoup
import requests

def get_top_links(query):
    # Search for the query and get top 5 links
    top_links = []
    for j in search(query, num=5, stop=5):
        top_links.append(j)
    return top_links

def main():
    name = input("Enter the celebrity name: ")
    links = get_top_links(name)
    print(f"Top 5 links for {name}:")
    for link in links:
        print(link)

if __name__ == "__main__":
    main()
```
