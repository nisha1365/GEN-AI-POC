```python
df['author_info'] = df['author_info'].astype(str)

# Extract 'id', 'name', and 'username' into separate columns
df['id'] = df['author_info'].str.extract(r"'id':\s*'(\d+)'", expand=False)
df['name'] = df['author_info'].str.extract(r"'name':\s*'([^']+)'", expand=False)
df['username'] = df['author_info'].str.extract(r"'username':\s*'([^']+)'", expand=False)

print(df)

```

```python
new_df = df[['author_id', 'id','name','username']].copy()
new_df.to_csv('Author_id_info.csv', index = False)
```
