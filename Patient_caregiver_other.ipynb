```python
import pandas as pd
```

```python
df = pd.read_csv("/content/model_generated_output_29th_march (1).csv")
```

```python
patient_seeds = [
    'patient', 'treatment', 'diagnosis', 'my condition', 'surgery', 'therapy',
    'chemotherapy', 'symptom', 'medication', 'doctor', 'hospital', 'recovery',
    'illness', 'disease', 'injection', 'side effect', 'nurse', 'healthcare',
    'medical', 'clinic', 'prescription', 'diagnosed', 'health condition',
    # Add more seeds to reach 50+
]

caregiver_seeds = [
    'caregiver', 'caring', 'look after', 'supporting', 'nursing', 'my patient',
    'care for', 'assist', 'helping', 'home care', 'caregiving', 'assistance',
    'support system', 'aid', 'family care', 'elder care', 'patient care',
    'health aide', 'care plan', 'daily care', 'care team', 'care needs',
    # Add more seeds to reach 50+
]

# Define a function to categorize tweets based on seed words
def categorize_tweet(tweet, patient_seeds, caregiver_seeds):
    tweet_lower = tweet.lower()
    if any(seed in tweet_lower for seed in patient_seeds):
        return 'Patient'
    elif any(seed in tweet_lower for seed in caregiver_seeds):
        return 'Caregiver'
    else:
        return 'Other'

# Apply the categorization function to your DataFrame
df['category'] = df['text'].apply(lambda x: categorize_tweet(x, patient_seeds, caregiver_seeds))

# Extract tweets based on the category
patients_tweets = df[df['category'] == 'Patient']
caregivers_tweets = df[df['category'] == 'Caregiver']
others_tweets = df[df['category'] == 'Other']
```
